* Exercici 3 : Crea una matriu bidimensional amb una forma de 5 x 5. Extreu el valor màxim de la matriu, i els valors màxims de cadascun dels seus eixos.

```python
import random

dim = 5
random_list = []
# Set a length of the list to x
for i in range(0, dim*dim):
    # any random numbers from 0 to ndim*ndim
    random_list.append(random.randint(0, 10))
print('List : ',random_list)
data = np.array(random_list).reshape(dim, dim)
print('Array :\n{}\n'.format(data))
max_arr = np.amax(data)         # Maximum of the flattened array
print('Maximum of array : ', max_arr)
max_col = np.amax(data, axis = 0)   # Maxima along the first axis
print('Maximum along the first axis : ', max_col)
max_row = np.amax(data, axis = 1)   # Maxima along the second axis
print('Maximum along the second axis : ', max_row)
```

Mostreu-me amb exemples de diferents matrius, la regla fonamental de Broadcasting que diu : "les matrius es poden transmetre / broadcast si les seves dimensions coincideixen o si una de les matrius té una mida d'1".

```python
# Example 1 : Single Dimension array

a = np.array([17, 11, 19]) # 1x3 Dimension array

print('vector \'a\': ', a)
b = 3  
print('scalar \'b\': ', b)
  
# Broadcasting happened beacuse of miss match in array Dimension.
c = a + b 
print('addition \'a + b\': ', c)


# Example 2 : Two Dimensional Array

M1 = np.array([[3, 6, 9], [5, -10, 15], [-7, 14, 21]])
M2 = np.array([[9, -18, 27], [11, 22, 33], [13, -26, 39]])
print('Matrix \'M1\'\n{}\n'.format(M1))
print('Matrix \'M2\'\n{}\n'.format(M2))

# Matrix Addition: To perform addition on the matrix, we will create two matrices
# using numpy.array() and add them using the (+) operator.
M3 = M1 + M2
print('Matrix addition: \'M1 + M2\'\n{}\n'.format(M3))

# Matrix Subtraction: To perform subtraction on the matrix, we will create two matrices
# using numpy.array() and subtract them using the (-) operator.
M3 = M1 - M2  
print('Matrix Subtraction: \'M1 - M2\'\n{}\n'.format(M3))

# Matrix dot product: First will create two matrices using numpy.array(). To multiply them will,
# you can make use of numpy dot() method. Numpy.dot() is the dot product of matrix M1 and M2.
# Numpy.dot() handles the 2D arrays and perform matrix multiplications.
M3 = np.array([[3, 6], [5, -10]])
M4 = np.array([[9, -18], [11, 22]])
print('Matrix \'M3\':\n{}\n'.format(M3))
print('Matrix \'M4\':\n{}\n'.format(M4))

M5 = M3.dot(M4)  
print('Matrix dot product \'M3 \u2219 M4\':\n{}\n'.format(M5))

# Matrix Multiplication:
M6 = np.random.randint(5, size=(3,2))
M7 = np.random.randint(5, size=(3,2))
#M7 = np.random.randint(5, size=(2,3))
print('Matrix \'M6\':\n{}\n'.format(M6))
print('Matrix \'M7\':\n{}\n'.format(M7))

# Matrix transposition
#M7 = M7.T
M7 = M7.transpose()
print('Matrix Transpose \'[M7]\u1d40\':\n{}\u1d40\n'.format(M7))

M8 = np.dot(M6,M7)
print('Matrix Multiplication \'M3 \u00d7 M4\':\n{}\n'.format(M5))

# Matrix Transpose: The transpose of a matrix is calculated, by changing the rows as columns and columns as rows.
# The transpose() function from Numpy can be used to calculate the transpose of a matrix.
M6 = np.array([[3, 6, 9], [5, -10, 15], [4,8,12]])
print('Matrix \'M6\':\n{}\n'.format(M6))
#M7 = M6.transpose()
M7 = M6.T
print('Matrix Transpose \'[M7]\u1d40\':\n{}\u1d40\n'.format(M7))
```

```python
# Format BGR
# If X is an array, it can have the following shapes:
# MxN – luminance (grayscale, float array only)
# MxNx3 – RGB (float or uint8 array)
# MxNx4 – RGBA (float or uint8 array)

img = np.asarray(Image.open('Tiger1.png')).copy()
img.setflags(write=1)
#print(img.flags)
#print(img)
#print(img.shape)
img1 = img.copy()
img2 = img.copy()
img3 = img.copy()

# Original Image
fig = plt.figure(figsize=(15,15))
ax = fig.add_subplot(2, 2, 1)
imgplot = plt.imshow(img)
ax.set_title('Original Image RGB')

# Image without Red Color 
img1[:,:,0] = 0 # Red
ax = fig.add_subplot(2, 2, 2)
imgplot = plt.imshow(img1)
imgplot.set_clim(0.0, 2.0)
ax.set_title('Image without Red Color')

# Image without Green Color 
img2[:,:,1] = 0 # Green
ax = fig.add_subplot(2, 2, 3)
imgplot = plt.imshow(img2)
imgplot.set_clim(0.0, 2.0)
ax.set_title('Image without Green Color')

# Image without Blue Color 
img3[:,:,2] = 0 # Blue
ax = fig.add_subplot(2, 2, 4)
imgplot = plt.imshow(img3)
imgplot.set_clim(0.0, 2.0)
ax.set_title('Image without Blue Color')

plt.savefig('Tiger1_cp.png')
#plt.imsave('Tiger1_cp.png', arr = img,format='png')
```
